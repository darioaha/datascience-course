```python
import pandas as pd
import numpy as np
import re

df = pd.read_csv("../Data/sales.csv"
           , encoding = 'latin1')
print(df.dtypes)
df.head()
```

```python
def sumar_uno( val ):
    try:
        res = int(val)
        return res+1
    except ValueError:
        print("Value ",val, "is not a number")
        
df.Account = df.Account.apply(sumar_uno)
#df2 = df.apply(sumar_uno, index=['Account']), axis=1)
df
```

```python
# Cambiar todos los valores de la columna price mayores a 50000 por NaN. 
def nanif_mayor_50000(val):
    nan=float('nan')
    try:
        if (val > 50000):
            return nan
        return val
    except ValueError:
        print("Value ",val, "is not a number")
    
print("==========APPLY NAN IF > 50000===============") 
df2 = df.Price.apply(nanif_mayor_50000)

# Luego, rellenar estos valores con el anterior en la columna.
mask_is_nan = df2.isna()
df2[mask_is_nan]

print("============COUNT BY ACCOUNT=================")
df.Account.value_counts()
```
