```python
def camel_to_snake(column_name):
    """
    Converts a string that is camelCase into snake_case
    """
    s1 = re.sub('(.)([A-Z][a-z]+)', r'\1_\2', column_name)
    return re.sub('([a-z0-9])([A-Z])', r'\1_\2', s1).lower()

#Define Data

df.columns = [camel_to_snake(col) for col in df.columns]
df.sample(3)
```

### Generar histogramas
Genera histogramas de datos numericos dividiendo por cantidad de bins

```python
def histogram1(variable, df, color, bins):
    fig = df[variable].hist(bins=bins, color=color)
    fig.set_xlabel(variable)
    fig.set_ylabel('Cantidad de Registros')
    fig.set_title('Distribución de '+variable)
    plt.draw()
    plt.savefig("graph/histogramas/"+variable+"_histogram1_"+str(bins)+".jpg")
    plt.clf()
```

### Discretizar variables en quartiles o con la mediana

```python
def make_discrete_df(variable, df, option=4):
    """Esta funcion toma una variable, df y opcion, y retorna el df original con la nueva variable discretizada
        Options:
        2 	= Median
        4 	= Quartiles
    """
    OPT = str(option)
    VAR = str(variable)
    VAR_discrete = VAR+'_discrete'+OPT
    data = df
    data.columns = [camel_to_snake(col) for col in data.columns]
    data.rename(columns={VAR: VAR_discrete}, inplace=True)

    #Median
    if option == 2:
        median = pd.qcut(data[VAR_discrete], 2, labels=['0', '1'])
        data.rename(columns={VAR_discrete: VAR}, inplace=True)
        cols_to_keep = data.columns
        data_new = data[cols_to_keep].join(median)
        return data_new
    #Quartiles
    elif option == 4:
        quantiles = pd.qcut(data[VAR_discrete], 4, labels=['1', '2', '3', '4'])
        data.rename(columns={VAR_discrete: VAR}, inplace=True)
        cols_to_keep = data.columns
        data_new = data[cols_to_keep].join(quantiles)
        return data_new
```

```python
def plot_pie(column,df_regular,df_not_regular):
    trace1 = go.Pie(values  = df_regular[column].value_counts().values.tolist(),
                    labels  = df_regular[column].value_counts().keys().tolist(),
                    hoverinfo = "label+percent+name",
                    domain  = dict(x = [0,.48]),
                    name    = "Regulares",
                    marker  = dict(line = dict(width = 2,
                                               color = "rgb(243,243,243)")
                                  ),
                    hole    = .6
                   )
    trace2 = go.Pie(values  = df_not_regular[column].value_counts().values.tolist(),
                    labels  = df_not_regular[column].value_counts().keys().tolist(),
                    hoverinfo = "label+percent+name",
                    marker  = dict(line = dict(width = 2,
                                               color = "rgb(243,243,243)")
                                  ),
                    domain  = dict(x = [.52,1]),
                    hole    = .6,
                    name    = "No regulares" 
                   )
    layout = go.Layout(dict(title = column + " distribucion de alumnos no regulares ",
                            plot_bgcolor  = "rgb(243,243,243)",
                            paper_bgcolor = "rgb(243,243,243)",
                            annotations = [dict(text = "Alumnos regulares",
                                                font = dict(size = 13),
                                                showarrow = False,
                                                x = .15, y = .5),
                                           dict(text = "Alumnos no regulares",
                                                font = dict(size = 13),
                                                showarrow = False,
                                                x = .88,y = .5
                                               )
                                          ]
                           )
                      )
    data = [trace1,trace2]
    fig  = go.Figure(data = data,layout = layout)
    return fig
    py.iplot(fig)
```
