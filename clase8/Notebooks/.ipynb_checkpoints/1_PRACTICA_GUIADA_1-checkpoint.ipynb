### ¿Como podriamos discernir los elementos contenidos en  2nd?

### Antes profundicemos un poco sobre los conceptos de Matplotlib.


**Figure**: es el espacio donde vamos a graficar, puede contener uno o más de un Axes (parcelas). Podemos pensar en una figura como un lienzo que contiene gráficos.


**Axes**: Es lo que generalmente consideramos como un gráfico (plot). Una figura puede contener muchos axes. A su vez cada Axes contiene dos o tres (en el caso de 3D) objetos Axis (ejes). Cada Axes tiene un título, una etiqueta x y una etiqueta y.

**Axis**: son los ejes de cada gráfico

Podemos crear una figura con el método plt.figure(), pero si no lo hacemos explícitamente, Matplotlib va a generar una figura automáticamente y en la figura va a crear el espacio para un gráfico. Podemos modificar la composición del lienzo con el método plt.subplots()

```python
fig, ax = plt.subplots(3,2, figsize=(10,7))

print('Type de fig: ', type(fig))
print('\n')
print('Type de ax: ', type(ax))
print('\n')
print('ax: ', ax)
```

Veamos ahora cómo crear un segundo eje en el mismo gráfico:

```python
# creamos la Figure "fig" y el AxesSubplot "ax1"
fig, ax1 = plt.subplots()

ax1.plot(y[:, 0], 'b', lw=1.5, label='1st')
ax1.plot(y[:, 0], 'ro')

ax1.grid(True)
ax1.legend(loc=0)
ax1.axis('tight')
ax1.set_xlabel('index')  # ojo: el método del AxesSubplot es .set_xlabel()
ax1.set_ylabel('value 1st')
ax1.set_title('A Simple Plot')

# inicializacion de un nuevo eje y con el método twinx de twin axis:
ax2 = ax1.twinx()

ax2.plot(y[:, 1], 'g', lw=1.5, label='2nd')
ax2.plot(y[:, 1], 'ro')
ax2.legend(loc=0)
ax2.set_ylabel('value 2st NUEVA INSTANCIA!')
ax2.grid(False)
```

También podemos trabajar como lo veníamos haciendo, invocando los métodos directamente desde plt. En este caso, los métodos aplican para el último axes inicializado.

```python
fig, ax1 = plt.subplots()

plt.plot(y[:, 0], 'b', lw=1.5, label='1st')
plt.plot(y[:, 0], 'ro')

plt.grid(True)
plt.legend(loc=8)
plt.axis('tight')
plt.xlabel('index')
plt.ylabel('value 1st')
plt.title('A Simple Plot')

# inicializacion de un nuevo eje y con el método twinx de twin axis:
ax2 = ax1.twinx()

plt.plot(y[:, 1], 'g', lw=1.5, label='2nd')
plt.plot(y[:, 1], 'ro')
plt.legend(loc=0)
plt.ylabel('value 2st NUEVA INSTANCIA!')
plt.grid(False)
```

###### Tambien existe la posibidad de hacer dos graficos alineados:

## Anotaciones sobre los gráficos

### Posiciones del texto

A veces es necesario anclar el texto a una posición en los ejes o en la figura, de forma independiente de los datos. 

En Matplotlib esto se logra modificando la transformación de los datos. 

Para graficar es necesario traducir entre sistemas de coordenadas. Un punto de datos (x,y) debe ser representado de alguna manera en una cierta ubicación en la figura, que a su vez debe ser representada en pixels en la pantalla. 

Existen 3 transformaciones predefinidas que pueden ser útiles. 

**ax.transData**: transformación asociada con las coordenadas de datos. 

**ax.transAxes**: transformación asociada con los Axes (en unidades de dimensiones de los ejes)

**fig.transFigure**: transformación asociada con la figura  (en unidades de dimensiones de la figura)

```python
fig, ax = plt.subplots(facecolor='lightgray')
ax.axis([0, 10, 0, 10])

# Asociado con las coordenadas de los datos, en las unidades de medida expresadas por los ejes
ax.text(1, 5, ". Data: (1, 5)", transform = ax.transData)

# Asociado con los ejes, en proporciones de los ejes: 
#(0,0) es el limite inferior izquierdo y (1,1) el superior derecho
ax.text(0.5, 0.1, ". Axes: (0.5, 0.1)", transform=ax.transAxes)

# Asociado con la figura
ax.text(0.2, 0.2, ". Figure: (0.2, 0.2)", transform=fig.transFigure);
```

#### Por default el texto es alineado por encima y a la izquieda de las coordenadas especificadas. El "." al comienzo de cada string  marcará aproximadamente la ubicación de la coordinada indicada.


#### Cambiamos los límites de los ejes y sólo las coordinadas de  transData se modifican. 

```python
ax.set_xlim(0, 2)
ax.set_ylim(-6, 6)
fig
```

Un ejemplo más con varios subplots:

```python
fig = plt.figure(facecolor='lightgray', figsize=(10,7))

ax = plt.subplot(222)

ax.axis([0, 10, 0, 10])

# Asociado con las coordenadas de los datos, en las unidades de medida expresadas por los ejes
ax.text(1, 5, ". Data: (1, 5)", transform = ax.transData)

# Asociado con los ejes, en proporciones de los ejes: 
#(0,0) es el limite inferior izquierdo y (1,1) el superior derecho
ax.text(0.5, 0.1, ". Axes: (0.5, 0.1)", transform=ax.transAxes)

# Asociado con la figura
ax.text(0.2, 0.2, ". Figure: (0.2, 0.2)", transform=fig.transFigure);
```

## Flechas y anotaciones

```python
fig, ax = plt.subplots()

x = np.linspace(0, 20, 1000)
ax.plot(x, np.cos(x))
ax.axis('equal')

ax.annotate('local maximum', xy=(2*np.pi, 1), xytext=(10, 4),
            arrowprops={'facecolor':'black', 'shrink': 0.5})

ax.annotate('local minimum', xy=(5 * np.pi, -1), xytext=(2, -6),
            arrowprops = {'arrowstyle' : '->',
                          'connectionstyle' : 'angle3,angleA=0,angleB=90'});
```

### Veamos un ejemplo

```python
import pandas as pd
from datetime import datetime
```

```python
data = pd.read_csv('../Data/spx.csv', index_col=0, parse_dates=True)
```

```python
spx = data['SPX']
```

```python
spx.head()
```

```python
fig, ax = plt.subplots(figsize=(12, 4))
spx.plot(ax=ax, style='k-');
```

```python
crisis_data = [
    (datetime(2007, 10, 11), 'Peak of bull market'),
    (datetime(2008, 3, 12), 'Bear Stearns Fails'),
    (datetime(2008, 9, 15), 'Lehman Bankruptcy')
]
```

```python
fig, ax = plt.subplots(figsize=(12, 4))
spx.plot(ax=ax)

for date, label in crisis_data:
    ax.annotate(label, xy=(date, spx.loc[date] + 75),
                xytext=(date, spx.loc[date] + 225),
                arrowprops=dict(facecolor='black', headwidth=4, width=2,
                                headlength=4),
                horizontalalignment='left', verticalalignment='top')

# Zoom in on 2007-2010
ax.set_xlim(['1/1/2007', '1/1/2011'])
ax.set_ylim([600, 1800])

ax.set_title('Important dates in the 2008-2009 financial crisis');
```

```python
from matplotlib.patches import Polygon
def func(x):
    return 0.5 * np.exp(x) + 1

a, b = 0.5, 1.5  # límites de la integral
x = np.linspace(0, 2)
y = func(x)

fig, ax = plt.subplots(figsize=(7, 5))
plt.plot(x, y, 'b', linewidth=2)
plt.ylim(ymin=0)

# Grafiquemos el valor de la integral, el aréa debajo de la función entre el límite superior e inferior 
Ix = np.linspace(a, b)
Iy = func(Ix)
verts = [(a, 0)] + list(zip(Ix, Iy)) + [(b, 0)]
poly = Polygon(verts, facecolor='0.7', edgecolor='0.5')
ax.add_patch(poly)

plt.text(0.5 * (a + b), 1, r"$\int_a^b f(x)\mathrm{d}x$",
         horizontalalignment='center', fontsize=20)

plt.figtext(0.9, 0.075, '$x$')
plt.figtext(0.075, 0.9, '$f(x)$')

ax.set_xticks((a, b))
ax.set_xticklabels(('$a$', '$b$'))
ax.set_yticks([func(a), func(b)])
ax.set_yticklabels(('$f(a)$', '$f(b)$'))
plt.grid(True)
#Función exponencial, aréa debajo de la integral
```

## Plots en 3D

```python
strike = np.linspace(50, 150, 24)
ttm = np.linspace(0.5, 2.5, 24)
strike, ttm = np.meshgrid(strike, ttm)
```

```python
print('strike.shape = {}'.format(strike.shape))
print('ttm.shape = {}'.format(ttm.shape))
```

```python
iv = (strike - 100) ** 2 / (100 * strike) / ttm
#volatilidades implícitas simuladas
```

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(9, 6))
ax = fig.gca(projection='3d')

surf = ax.plot_surface(strike, ttm, iv, rstride=2, cstride=2,
                       cmap=plt.cm.coolwarm, linewidth=0.5,
                       antialiased=True)

ax.set_xlabel('strike')
ax.set_ylabel('time-to-maturity')
ax.set_zlabel('implied volatility')

fig.colorbar(surf, shrink=0.5, aspect=5);

# 3D surface plot
```

```python
fig = plt.figure(figsize=(8, 5))
ax = fig.gca(projection='3d')
ax.view_init(30, 60)

ax.scatter(strike, ttm, iv, zdir='z', s=25,
           c='b', marker='^')

ax.set_xlabel('strike')
ax.set_ylabel('time-to-maturity')
ax.set_zlabel('implied volatility');

# 3d scatter plot
```

```python
# Podemos ver los dos gráficos juntos con subplots:

fig = plt.figure(figsize=(16, 5))
ax = fig.add_subplot(121, projection='3d')
ax1 = fig.add_subplot(122, projection='3d')

ax.scatter(strike, ttm, iv, zdir='z', s=25,
           c='b', marker='^')

ax.view_init(30, 60)
ax.set_xlabel('strike')
ax.set_ylabel('time-to-maturity')
ax.set_zlabel('implied volatility')


ax1.plot_surface(strike, ttm, iv, rstride=2, cstride=2,
                       cmap=plt.cm.coolwarm, linewidth=0.5,
                       antialiased=True)

ax1.view_init(30, 60)
ax1.set_xlabel('strike')
ax1.set_ylabel('time-to-maturity')
ax1.set_zlabel('implied volatility');
```
